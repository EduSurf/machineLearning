## Criando novas variáveis

```python
import pandas as pd
```

```python
uri = 'C:\\Users\\F5440952\\Projetos Python\\Curso Pandas\\dados\\aluguel_residencial.csv'
dados = pd.read_csv(uri, sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(10)
```

```python
dados['Valor bruto m2'] = (dados['Valor bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo variaveis do dataframe

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor bruto', 'Valor bruto m2']])
dados_aux.head(10)
```

```python
del dados_aux['Valor bruto']
dados_aux.head(10)
```

```python
dados_aux.pop('Valor bruto m2')
dados_aux.head(10)
```

```python
dados.drop(['Valor bruto', 'Valor bruto m2'], axis = 1, inplace = True)
dados.head(10)
```

```python
dados.to_csv('C:\\Users\\F5440952\\Projetos Python\\Curso Pandas\\dados\\aluguel_residencial.csv', sep = ';', index = False)
```
