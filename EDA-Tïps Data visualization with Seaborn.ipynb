## Teste de hipótese

### H<sup>null</sup>

### A distribuição da taxa de gorjeta é a mesma nos dois grupos

### H<sup>alt</sup>

### A distribuição da taxa de gorjeta não é a mesma nos dois grupos

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
sobremesa
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
sem_sobremesa
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print('O valor P é {}'.format(r.pvalue))
# significa que a probabilidade das amostras representarem a população é alta, de 0.5266/1. Caso seja abaixo de 0.05/1,
# a hipótese alternativa deveria ser considerada
```

## Teste de hipótese

## H<sup>null</sup>

## A distribuição do valor da conta é igual no sábado e no domingo

## H<sup>alt</sup>

## A distribuição do valor da conta é igual no sábado e no domingo

```python
gorjetas.head()
```

```python
gorjetas = gorjetas.rename(columns = {'dia da semana' : 'dia_da_semana'})
valor_conta_domingo = gorjetas.query("dia_da_semana == 'Domingo'")['Valor da conta']
valor_conta_sabado = gorjetas.query("dia_da_semana == 'Sábado'")['Valor da conta']
```

## p value > 0.05 aceita H<sup>null</sup>, cc aceita H<sup>alt</sup> 

```python
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
print('O valor do P-Value é {}'.format(r2.pvalue))
```
